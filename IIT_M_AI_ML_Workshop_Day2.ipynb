**MACHINE LEARNING**

1.  It is a process of making computer human intelligence

**TYPES OF MACHINE LEARNING**

1.  Supervised Learning ( Knowing the input - trying to predict the output )
2.  Unsupervised Learning ( Unkown input trying to predict the output )
3.  Reinforcement Learning ( Feedback based learning )

**Problem in learning**
- Regression (continuous flow of values )
- Classification ( Classifying classes )
- Cluster ( Grouping of identical items )

**TYPES OF REGRESSION**
- simple linear regression (y=mx+c, y is dependent variable, x is independent variable )
- multiple linear regression (y=m1+m2+m3+.........+c)
- polynomial linear regression(y=m1x1+m2x^2+m3x^3+c)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
import seaborn as sb
```

```python
house={"Area":[5,3,4,6,7,5],"Price":[7,4,6,9,3,2]}
```

```python
data=pd.DataFrame(house)
data
```

```python
data.describe()
```

```python
num,den = 0,0
for i in range(len(data)):
  num = num +(data['Area'][i]-np.mean(data.Area))*(data['Price']-np.mean(data.Price))
  den = den+(data['Area'][i]-np.mean(data.Area))**2
num/den
```

**WHY REGRESSION ??**


```python
pt.plot(data.Area, data.Price) #the below graph is not a straight line in order to make it a straight line use regression model
```

```python
x=data[["Area"]]
y=data["Price"]
```

```python
## applying regression model now
```

```python
from sklearn.linear_model import LinearRegression

#sklearn = scikit learn (Module name is scipy)
```

```python
thanos=LinearRegression()
type(thanos)
```

```python
thanos.fit(x,y)
```

```python
thanos.coef_ #slope value
```

```python
thanos.intercept_
```

```python
thanos.predict([[100]]) #price prediction of the particular area
```

```python
shiv = thanos.predict(x)
```

```python
pt.plot(data.Area,shiv)
pt.show()
```

```python
pt.plot(data.Area,shiv)
pt.scatter(data.Area,shiv)
pt.scatter(data.Area,data.Price)
pt.show()
```

**PERFORMANCE ANALYSIS OF THE MODEL**

```python
mean_absolute_error(y,shiv)
```

```python
mean_squared_error(y,shiv)
```

```python
np.sqrt(mean_squared_error(y,shiv))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y,shiv)*100  #checks the accuracy
```
